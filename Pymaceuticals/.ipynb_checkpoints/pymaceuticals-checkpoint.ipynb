# Observations and Insights 


Ramicane drug is the most effective followed by Capomulin. 

Gender of the mouse doesnt show any significant role in the Data set.

Infubinol is the only drug among the four, which is having one outlier. 

Capomulin drug seems to be effective and the data show decrease in tumor volume over the period of the study.

There is positive correlation between the weight of the mouse and average tumor volume in the study of Capomulin. 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
from scipy.stats import linregress
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
merge_df = pd.merge(mouse_metadata, study_results, on="Mouse ID", how="outer")
# Display the data table for preview
merge_df
```

```python
# Checking the number of mice.
len(merge_df["Mouse ID"].unique())
```

```python
#Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint.

# To find the duplicate mice ID
merge_df[merge_df.duplicated(["Mouse ID","Timepoint"])]

# Find all the duplicate data for the above mice ID
duplicate = merge_df[merge_df["Mouse ID"] == "g989"]
duplicate
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.

clean = merge_df[merge_df["Mouse ID"] != 'g989']
clean
```

```python
# Checking the number of mice in the clean DataFrame.
len(clean["Mouse ID"].unique())
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
group_regimen = clean.groupby("Drug Regimen")

group_mean = group_regimen["Tumor Volume (mm3)"].mean()
group_median = group_regimen["Tumor Volume (mm3)"].median()
group_variance = group_regimen["Tumor Volume (mm3)"].var()
group_sdeviation = group_regimen["Tumor Volume (mm3)"].std()
group_sem = group_regimen["Tumor Volume (mm3)"].sem()
group_sem

# This method is the most straighforward, creating multiple series and putting them all together at the end.

summary_clean = pd.DataFrame({"Mean": group_mean, "Median": group_median, "Variance": group_variance, "Standard Deviation": group_sdeviation, "SEM": group_sem})
summary_clean
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# This method produces everything in a single groupby function
data = clean.groupby("Drug Regimen").agg({"Tumor Volume (mm3)":["mean","median","var","std","sem"]})
data.sort_index()
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  

clean

# # Capomulin, Ramicane, Infubinol, and Ceftamin

# # Start by getting the last (greatest) timepoint for each mouse

# # Capomulin
capomulin_df = clean[clean["Drug Regimen"] == "Capomulin"]
capomulin_df_max=capomulin_df.drop_duplicates("Mouse ID", keep="last")
capomulin_df_max

# # # Ramicane
ramicane_df = clean[clean["Drug Regimen"] == "Ramicane"]
ramicane_df_max=ramicane_df.drop_duplicates("Mouse ID", keep="last")
ramicane_df_max

# # # Infubinol
infubinol_df = clean[clean["Drug Regimen"] == "Infubinol"]
infubinol_df_max=infubinol_df.drop_duplicates("Mouse ID", keep="last")
infubinol_df_max

# # # Ceftamin
ceftamin_df = clean[clean["Drug Regimen"] == "Ceftamin"]
ceftamin_df_max=ceftamin_df.drop_duplicates("Mouse ID", keep="last")
ceftamin_df_max


# # # Merge this group df with the original dataframe to get the tumor volume at the last timepoint

merge_drug = pd.concat([capomulin_df_max, ramicane_df_max,infubinol_df_max,
                        ceftamin_df_max]).sort_values(["Tumor Volume (mm3)"],ascending = True).reset_index()
merge_drug
```

```python
drug_list = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# Create empty list to fill with tumor vol data (for plotting)

tumor_volumes = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
# Locate the rows which contain mice on each drug and get the tumor volumes

for drug in drug_list:
    tumor_vol_by_drug = merge_drug['Tumor Volume (mm3)'].loc[merge_drug['Drug Regimen'] == drug]
    tumor_volumes.append(tumor_vol_by_drug)
    
# Determine outliers using upper and lower bounds

    quartiles = tumor_vol_by_drug.quantile([.25,.5,.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq-lowerq
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    outliers= tumor_vol_by_drug.loc[tumor_vol_by_drug< lower_bound]
    
    
    print(f"{drug}'s potential outliers: {outliers}")
```

```python
 # Generate a box plot of the final tumor volume of each mouse across four regimens of interest

# format the outliers differently


# plot
fig, ax1 = plt.subplots()
ax1.set_xticklabels(drug_list)
ax1.set_ylabel('Final Tumor Volume (mm3)')
ax1.boxplot(tumor_volumes,flierprops={'marker': 'o', 'markersize': 10, 'markerfacecolor': 'Red'})
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of time point versus tumor volume for a mouse treated with Capomulin

mouse = clean[clean["Mouse ID"] == "l509"]
x_axis = mouse["Timepoint"]
y_axis = mouse["Tumor Volume (mm3)"]

plt.plot(x_axis,y_axis)
plt.title("Capomulin treatment of mouse l509")
plt.xlabel("Timepoint(days)")
plt.ylabel("Tumor Volume (mm3)")
plt.show()
```

```python
# Generate a scatter plot of mouse weight versus average tumor volume for the Capomulin regimen
camp_drug = clean[clean["Drug Regimen"] == "Capomulin"]
camp_drug
unique_mouse = camp_drug["Mouse ID"].unique()

weight = []

average = []

for mouse in unique_mouse:
    tumor_average = clean.loc[clean["Mouse ID"] == mouse, "Tumor Volume (mm3)"].mean()
    average.append(tumor_average)
    mouse_weight = clean.loc[clean["Mouse ID"] == mouse, "Weight (g)"].mean()
    weight.append(mouse_weight)

plt.scatter(weight,average)
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.title("Weight Vs. Average Tumor Volume(Capumulin)")


plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen

print(f"The correlation between mouse weight and the average tumor volume is {round(st.pearsonr(weight,average)[0],2)}")

```

```python
plt.scatter(weight,average)
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.title("Weight Vs. Average Tumor Volume(Capumulin)")



(slope, intercept, rvalue, pvalue, stderr) = linregress(weight, average)

# Calculate regression value
reg_values = np.asarray(weight) * slope + intercept

# Calculating Line equation
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

# plotting regression line on scatter plot

plt.plot(weight,reg_values,"r-")
plt.annotate(line_eq,(20,45),fontsize=15,color="red")
plt.show()

```
