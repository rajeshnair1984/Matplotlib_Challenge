## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
merge_df = pd.merge(mouse_metadata, study_results, on="Mouse ID", how="outer")
# Display the data table for preview
merge_df
```

```python
# Checking the number of mice.
unique = merge_df["Mouse ID"].value_counts()
len(unique)
```

```python
#Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
# duplicate = merge_df.loc[merge_df.duplicated(subset=['Mouse ID',"Timepoint"]),"Mouse ID"]
# duplicate
duplicate = merge_df[merge_df[["Mouse ID","Timepoint"]].duplicated(keep=False)]
duplicate
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
drop_duplicate = duplicate.drop_duplicates(["Timepoint"])
drop_duplicate
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean= pd.concat([merge_df, drop_duplicate, drop_duplicate]).drop_duplicates(keep=False)
clean
```

```python
# Checking the number of mice in the clean DataFrame.
len(clean["Mouse ID"].unique())
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
group_regimen = clean.groupby("Drug Regimen")
group_regimen.count()
group_mean = group_regimen["Tumor Volume (mm3)"].mean()
group_median = group_regimen["Tumor Volume (mm3)"].median()
group_variance = group_regimen["Tumor Volume (mm3)"].var()
group_sdeviation = group_regimen["Tumor Volume (mm3)"].std()
group_sem = group_regimen["Tumor Volume (mm3)"].sem()
group_sem

# This method is the most straighforward, creating multiple series and putting them all together at the end.

```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# This method produces everything in a single groupby function

group_summary = pd.DataFrame({"Mean":group_mean,"Median":group_median,"Var":group_variance,
                             "S Dev":group_sdeviation,"SEM":group_sem})
group_summary
```

## Bar and Pie Charts

```python
# Generate a bar plot showing the total number of mice for each treatment throughout the course of the study using pandas. 
group_regimen = clean.groupby("Drug Regimen")
mice_count = group_regimen["Mouse ID"].count()
mice_count

mice_count.plot(kind="bar", color = 'blue',figsize=(13,5),fontsize = 12)
plt.title("Treatment Study On Mice",fontsize = 15)
plt.xlabel("Drug Regimen",fontsize = 12)
plt.ylabel("Number of Mice",fontsize = 12)
plt.show()
```

```python
# Generate a bar plot showing the total number of mice for each treatment throughout the course of the study using pyplot.
x_axis = mice_count.index.values
plt.figure(figsize=(13,5))
plt.bar(x_axis,mice_count,color="b", align="center")
plt.xlim(-.75, len(x_axis)-0.25)
plt.ylim(0, max(mice_count)+10)
plt.title("Treatment Study On Mice",fontsize = 15)
plt.xlabel("Drug Regimen",fontsize = 12)
plt.ylabel("Number of Mice",fontsize = 12)

plt.show()
```

```python
# Generate a pie plot showing the distribution of female versus male mice using pandas
clean

gender = clean[["Mouse ID","Sex"]]
gender

gender_duplicate = gender.drop_duplicates(["Mouse ID"])
gender_duplicate
gender_group = gender_duplicate.groupby("Sex")
count_gender = gender_group.count()
count_gender
count_gender.columns = ["Sex Count"]
count_gender
count_gender["Percentage Sex"] = (count_gender / count_gender.sum())*100
count_gender
count_gender.style.format({"Percentage Sex":"{:.2f}%"})

```

```python
explode = (0.1, 0)
gender_pie = count_gender.plot(kind="pie", y='Sex Count',startangle=140,explode=explode,
                               title=("Distribution of Female versus Male Mice"),autopct="%1.1f%%")
gender_pie.set_ylabel("Sex")
plt.axis("equal")
plt.show()
# plt.axis("equal")
```
